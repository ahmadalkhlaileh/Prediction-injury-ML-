```python
valid_values = ['Lean', 'Normal', 'Stocky']
for i in df['body_type']:
    if i not in valid_values:
        df.replace(i, 'Normal', inplace=True)
df["body_type"].unique()
```



1.   This is useful because **national_team** is a categorical feature
2.   **'value_euro'** and **release_clause_euro** because it is less affected by extreme outliers compared to the mean










```python
df['national_team'] = df['national_team'].fillna("None")

df['value_euro'] = df['value_euro'].fillna(df['value_euro'].median())

df['release_clause_euro'] = df['release_clause_euro'].fillna(df['release_clause_euro'].median())
```

```python
print(df['national_team'].isnull().sum(),
      df['value_euro'].isnull().sum(),
      df['release_clause_euro'].isnull().sum())
```

```python
df.isnull().sum()
```



1.   players without a national team association will have a default rating of 0
2.  players without a national team do not have a jersey number.



```python
df['wage_euro'] = df['wage_euro'].fillna(df['wage_euro'].median())

df['national_rating'] = df['national_rating'].fillna(0)

df['national_team_position'] = df['national_team_position'].fillna("None")

df['national_jersey_number'] = df['national_jersey_number'].fillna(0)
```

```python
print(df['wage_euro'].isnull().sum(),
      df['national_team_position'].isnull().sum(),
      df['national_rating'].isnull().sum(),
      df['national_jersey_number'].isnull().sum())
```



1.   from sklearn.preprocessing import MinMaxScaler (I CAN USE IT)
2.   normalizing the data can still help the optimization process by ensuring that all features contribute equally to the model's cost function.



```python
df['height_cm'] = (df['height_cm'] - df['height_cm'].min()) / (df['height_cm'].max() - df['height_cm'].min())
df['weight_kgs'] = (df['weight_kgs'] - df['weight_kgs'].min()) / (df['weight_kgs'].max() - df['weight_kgs'].min())
```

```python
df.isnull().sum()
```

i can use One Hot Encoding or Label Encoding

```python
df['preferred_foot'] = df['preferred_foot'].replace({'Left': 0, 'Right': 1})
df['preferred_foot']
```



*    One-hot encoding converts categories into binary columns.
*  Example: 'Color' ['Red', 'Blue', 'Green'] -> Red: [1, 0, 0], Blue: [0, 1, 0], Green: [0, 0, 1]



```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[['body_type']])

encoded_df = pd.DataFrame(encoded_features.toarray(), columns=encoder.get_feature_names_out(['body_type']))
df = pd.concat([df, encoded_df], axis=1)
```

Take first position for each player

```python
df['first_position'] = df['positions'].str.split(',').str[0]

def categorize_position(position):
    if position in ['CB', 'LB', 'RB', 'LWB', 'RWB']:
        return 'Defender'
    elif position in ['CM', 'CDM', 'CAM', 'LM', 'RM']:
        return 'Midfielder'
    elif position in ['ST', 'CF', 'LW', 'RW']:
        return 'Attacker'
    elif position == 'GK':
        return 'Goal keeper'
    else:
        return 'Unknown'


df['position_category'] = df['first_position'].apply(categorize_position)


df['position_category']
```

```python
encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[['position_category']])

encoder = OneHotEncoder()
encoded_features = encoder.fit_transform(df[['position_category']])

encoded_df = pd.DataFrame(encoded_features.toarray(), columns=encoder.get_feature_names_out(['position_category']))
df = pd.concat([df, encoded_df], axis=1)
```

Drop unnecessary columns

```python
columns_to_drop = [
    'first_position', 'positions', 'birth_date', 'name', 'full_name', 'national_team',
    'national_jersey_number', 'nationality', 'national_rating', 'national_team_position',
    'release_clause_euro', 'value_euro', 'wage_euro'
]
df = df.drop(columns=columns_to_drop)
```

Provide new colums to use it as (Target)


```python
df['injury'] = (
    (df['stamina'] < 50) |
    (df['strength'] < 50) |
    (df['balance'] < 50) |
    (df['jumping'] < 50) |
    (df['age'] > 30) |
    (df['aggression'] > 70) |
    ((df['aggression'] > 70) & (df['strength'] < 60)) |
    ((df['acceleration'] > 80) & (df['stamina'] < 40)) |
    ((df['sliding_tackle'] > 75) & (df['balance'] < 40))
)
df['injury'] = df['injury'].astype(int)
df['injury']
```

Creating interaction features

```python

df['age_stamina'] = df['age'] * df['stamina']
df['strength_aggression'] = df['strength'] * df['aggression']
```

 Creating polynomial features (squared terms)

*   could improve model performance, especially for algorithms sensitive to non-linear patterns (e.g., logistic regression or tree-based models).





```python

df['stamina_squared'] = df['stamina'] ** 2
df['strength_squared'] = df['strength'] ** 2
```



*   X (features): Everything except injury.

*   y (target): The injury column.

Importing necessary Libraries

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import cross_val_score

```

*  30% of the data will be used for testing, and 70% will be used for training.
*    **random_state** : ensures that the random split is reproducible
https://www.youtube.com/watch?v=LDQXXJphBlA


```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

1.   standardize the data.

2.   computes the mean and standard deviation of the training data and scales it


*must fit the scaler on the training data only, so that the model is not aware of the test data during training.*




```python
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

iterations ==> (1000) to ensure the algorithm converges



1.   scaled training data (X_train_scaled)  
2.   target variable (y_train).



```python
model = LogisticRegression(max_iter=1000, random_state=42)

model.fit(X_train_scaled, y_train)
```

```python
y_pred = model.predict(X_test_scaled)
```

# Evaluate Model Performance



*   Split the dataset {A, B, C, D, E} into 5 equal parts (or folds)


---

corss validation work :


1.   **Train on**: {B, C, D, E}, **Test on**: {A}


2.  **Train on**: {A, C, D, E}, **Test on** {B}




```python
accuracy = accuracy_score(y_test, y_pred)

conf_matrix = confusion_matrix(y_test, y_pred)

class_report = classification_report(y_test, y_pred)

val = cross_val_score(model, X_train_scaled, y_train, cv=5, scoring='accuracy')
```

```python
feature_names = X.columns


coefficients = model.coef_[0]


feature_importance = pd.DataFrame({
    'Feature': feature_names,
    'Coefficient': coefficients
})


feature_importance['Abs_Coefficient'] = feature_importance['Coefficient'].abs()
feature_importance = feature_importance.sort_values(by='Abs_Coefficient', ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x='Abs_Coefficient', y='Feature', data=feature_importance.head(15))
plt.title('Top 15 Important Features')
plt.xlabel('Absolute Coefficient Value')
plt.ylabel('Feature')
plt.show()
```

```python
print("Accuracy:", accuracy)
print("Cross-validation scores:" ,val)
```

# Why ( features with low or moderate correlation with the target variable still appear as highly important in certain models)?

***A low correlation between a feature and the target does not necessarily mean the feature is unimportant. Correlation is a simple tool, but feature importance often depends on the type of model, indirect relationships, or interactions between features.***






1.   Correlation measures **linear relationships**. If the relationship between a feature and the target is non-linear (e.g., quadratic or exponential), the correlation might be low even though the feature is important.

2.  Some features are only important in combination with others, which correlation cannot capture.


3.   If some features are highly correlated with each other, their individual correlation with the target might be Weak. The model might still assign importance to these features based on how they work together.
